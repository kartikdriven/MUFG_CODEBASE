# Introduction to Matplotlib

---

## 1. Theory

Matplotlib is a popular Python library for **data visualization**.  
It is widely used for plotting graphs and charts, allowing developers to represent data visually.  

Key features:
- Produces publication-quality plots.
- Supports line plots, bar charts, histograms, scatter plots, pie charts, etc.
- Can integrate with **NumPy** and **Pandas**.
- Provides control over figure size, labels, legends, colors, and styles.

---

##  Applications

- Visualizing statistical data.
- Understanding trends and patterns in datasets.
- Exploratory Data Analysis (EDA).
- Plotting real-time sensor data.
- Creating dashboards and reports.
- Used in scientific research, finance, engineering, and machine learning.

---


## List of Functions and Their Usage

| Function | Usage |
|----------|--------|
| `plt.plot(x,y)` | Line plot |
| `plt.bar(x,y)` | Bar chart |
| `plt.hist(data, bins)` | Histogram |
| `plt.scatter(x,y)` | Scatter plot |
| `plt.pie(values, labels)` | Pie chart |
| `plt.xlabel(), plt.ylabel()` | Set axis labels |
| `plt.title()` | Set plot title |
| `plt.legend()` | Add legend |
| `plt.grid()` | Show grid |
| `plt.subplot()` | Multiple plots in one figure |
| `plt.figure(figsize=(w,h))` | Define figure size |
| `plt.show()` | Display plot |


## Import and Execution

```python
# Importing Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Basic Line Plot
plt.plot(x, y, label="sin(x)", color="blue", linestyle="--")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Sine Wave Example")
plt.legend()
plt.grid(True)
plt.show()
```

## Examples

```python
# Example 1: Line Plot with Multiple Lines
import numpy as np

y2 = np.cos(x)
plt.plot(x, y, label="sin(x)", color="red")
plt.plot(x, y2, label="cos(x)", color="green")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sine and Cosine Waves")
plt.legend()
plt.show()

# Example 2: Bar Chart
categories = ['A','B','C','D']
values = [10, 24, 36, 18]
plt.bar(categories, values, color=['blue','orange','green','red'])
plt.title("Bar Chart Example")
plt.show()

# Example 3: Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.show()

# Example 4: Scatter Plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.scatter(x_scatter, y_scatter, color='darkgreen')
plt.title("Scatter Plot Example")
plt.show()

# Example 5: Pie Chart
sizes = [30, 20, 25, 25]
labels = ['Apple', 'Banana', 'Mango', 'Grapes']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()

# Example 6: Subplots
fig, axs = plt.subplots(2, 2, figsize=(8,8))

# Line Plot
axs[0,0].plot(x, y, color='blue')
axs[0,0].set_title("Line Plot")

# Bar Chart
axs[0,1].bar(categories, values, color='orange')
axs[0,1].set_title("Bar Chart")

# Histogram
axs[1,0].hist(data, bins=20, color='green')
axs[1,0].set_title("Histogram")

# Scatter Plot
axs[1,1].scatter(x_scatter, y_scatter, color='red')
axs[1,1].set_title("Scatter Plot")

plt.tight_layout()
plt.show()
```
